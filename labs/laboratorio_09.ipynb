El conjunto de datos se denomina `vehiculos_procesado_con_grupos.csv`, el cual contine algunas de las características más importante de un vehículo.

En este ejercicio se tiene como objetivo, es poder clasificar los distintos vehículos basados en las cracterísticas que se presentan a continuación. La dificultad de este ejercicio radíca en que ahora tenemos variables numéricas y variables categóricas.

Lo primero será cargar el conjunto de datos:

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.dummy import DummyClassifier
from sklearn.cluster import KMeans


%matplotlib inline
sns.set_palette("deep", desat=.6)
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","vehiculos_procesado_con_grupos.csv"), sep=",")\
       .drop(
            ["fabricante", 
             "modelo",
             "transmision", 
             "traccion", 
             "clase", 
             "combustible",
             "consumo"], 
    
          axis=1)

df.head()
```

En este caso, no solo se tienen datos numéricos, sino que también categóricos. Además, tenemos problemas de datos **vacíos (Nan)**. Así que para resolver este problema, seguiremos varios pasos:

## 1.- Normalizar datos

1. Cree un conjunto de datos con las variables numéricas, además, para cada dato vacía, rellene con el promedio asociado a esa columna. Finalmente, normalize los datos mediante el procesamiento **MinMaxScaler** de **sklearn**.

2.-  Cree un conjunto de datos con las variables categóricas , además, transforme de variables numéricas a categóricas ocupando el comando **get_dummies** de pandas ([refrenecia](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html)). Explique a grande rasgo como se realiza la codificación de variables numéricas a categóricas.

3.- Junte ambos dataset en uno, llamado **df_procesado**. 


```python
#Primero crearemos variables categóricas y continuas:

predictor_lucy = [] #variables continuas
predictor_chise = [] #variables categórica


for column in df.columns:
    if df[column].dtype == 'object':
        predictor_chise.append(column)
    else:
        predictor_lucy.append(column)
        
print(predictor_chise)
print(predictor_lucy)   
```

```python
 #tal como vimos, definiremos el dataframe como en clases

df_cont = df.drop(predictor_chise, axis=1)

for column in df_cont.columns: #reemplazamos los NaN por las medias respectivas de las columnas
    if df_cont[column].notnull().all() == False:
        df_cont = df_cont.fillna(df_cont[column].mean())
        
df_cont.head()
```

```python
scaler = MinMaxScaler() #normalizamos
df_cont[predictor_lucy] = scaler.fit_transform(df_cont[predictor_lucy])

df_cont.head()
```

```python
df_chise = df.drop(predictor_lucy, axis=1) #se crea el dataframe categórico, finalmente (como en el meme)
df_chise.head()
```

```python
df_chise_dum = pd.get_dummies(df_chise) #ponemos get_dummies para llevar valores cat a numéricos
df_chise_dum.head()
```

```python
df_procesa3 = df_cont.join(df_chise_dum) # definimos el dataframe procesado usando el comando join
df_procesa3.head()
```

```python
#finalmente, veremos la descripción de este nuevo dataframe

df_procesa3.describe()
```

## 2.- Realizar ajuste mediante kmeans

Una vez depurado el conjunto de datos, es momento de aplicar el algoritmo de **kmeans**.

1. Ajuste el modelo de **kmeans** sobre el conjunto de datos, con un total de 8 clusters.
2. Calcular los cluster y el valor de los centroides.
3. Realizar que resumas las principales cualidades de cada cluster. Para  cada cluster calcule:

    a. Valor promedio de las variables numérica.\
    b. Moda para las variables numericas
    
    

```python
# 1) la primera cosa que debemos hacer, es ajustar el modelo k-means:

from sklearn.cluster import KMeans #importamos la librería clásica

kane = KMeans()
kane.fit(df_procesa3)

centroids = kane.cluster_centers_ # centros 
clusters = kane.labels_ # clusters
```

```python
df_procesa3["cluster"] = clusters
df_procesa3["cluster"] = df_procesa3["cluster"].astype('category')
centroids_df_proc = pd.DataFrame(centroids, columns=df_procesa3.columns)
centroids_df_proc["cluster"] = [1,2,3,4,5,6,7,8]
```

```python
df_procesa3.head()
```

```python
df_procesa3['cluster'].unique() # los nuevos clusters
```

```python
#Se calcularan el a) Valor promedio de las variables numérica y b) Moda para las variables numericas

for i in range(1,9):
    print("Media de cluster ",i)
    print(df_procesa3[df_procesa3['cluster']==i-1][predictor_lucy].mean()) # media de cada cluster
    print("")
    print("Moda de cluster ",i)
    print(str(df_procesa3[df_procesa3['cluster']==i-1][predictor_lucy].mode())) # moda de cada cluster
    print("")
```

## 3.- Elegir Número de cluster

Estime mediante la **regla del codo**, el número de cluster apropiados para el caso.
Para efectos prácticos, eliga la siguiente secuencia como número de clusters a comparar:

$$[5, 10, 20, 30, 50, 75, 100, 200, 300]$$

Una ve realizado el gráfico, saque sus propias conclusiones del caso.


```python
# Implementaremos la mismísima regla del codo

chiseuwu = [5, 10, 20, 30, 50, 75, 100, 200, 300]
km = [KMeans(n_clusters=i) for i in chiseuwu]
sc = [km[i].fit(df_procesa3).inertia_ for i in range(len(km))]


df_codo = pd.DataFrame({'Number of Clusters':chiseuwu,
                        'Score':sc})

df_codo.head()
```

```python
#graficar todo con el k-means

fig, ax = plt.subplots(figsize=(11, 8.5))
plt.title('Curva de Codo')
sns.lineplot(x="Number of Clusters",
             y="Score",
            data=df_codo)
sns.scatterplot(x="Number of Clusters",
             y="Score",
             data=df_codo)
```
