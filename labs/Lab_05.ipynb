EL conjunto de datos se denomina `company_sales_data.csv`, el cual contiene información tal como: número del mes, unidades, precio, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
fig = plt.figure(figsize=(15,11))

#gráfico lineal
belloto = sns.lineplot(
    x='month_number',#eje x 
    y='total_profit',#eje y
    data=df, #datos
    ci = None,
    label='Gráfica Lineal',
)

#gráfico de dispersión
belloto = sns.scatterplot(
        x='month_number', #eje x
        y='total_profit', #eje y
        s=140, #tamaño del punto
        data=df, # datos
        color='r', # color
        label='dispersion',
)

plt.legend(loc='upper left', fontsize='x-large') 
plt.title('total_profit de todos los meses', fontsize=20) 
plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
#Ocuparemos un for in range, dentro de las columnas del datafram para realizar un gráfico multilínea.

fig = plt.figure(figsize=(15,11))

lulu = df.drop(['month_number', 'total_units', 'total_profit'], axis=1) # agrupacion de los datos
belloto = lulu.columns # columnas

for atun in belloto:
    sns.lineplot(
        x='month_number',
        y= atun,
        data=df,
        label=str(atun)
    )

plt.xlabel('Número del Mes', fontsize=20) # etiqueta eje X
plt.ylabel('Unidades Totales', fontsize=20) # etiqueta eje Y
plt.legend(loc='upper left', fontsize='x-large') # leyenda
plt.title('Ventas de todos los productos', fontsize=20) # titulo
plt.show()
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
#Al tener los mismos intervalos de números de meses en el eje X, podremos super poner los gráficos de barra:


empanadita = plt.figure(figsize=(14, 11))
empanadita = plt.bar(
        df['month_number'], # eje x
        df['facecream'], # eje y 
        align='center', # centrar nombre eje x
        color="k", # color 
        alpha=0.8, 
        label='facecream' # etiqueta
    )
empanadita = plt.bar(
        df['month_number'], #eje x
        df['facewash'], #eje y 
        align='center', #centrar nombre eje x
        color="r", #color 
        alpha=0.8,
        label='facewash' # etiqueta
    )

plt.legend(fontsize='large') # leyenda
plt.xlabel('Número del Mes', fontsize=30) # etiqueta eje X
plt.ylabel('Unidades Totales', fontsize=30) # etiqueta eje Y
plt.title('Ventas de facecream y facewash', fontsize=30) # titulo
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
fig = plt.figure(figsize=(15,11))
sns.boxplot(data=lulu) # grafico de caja
sns.swarmplot(data=lulu, color='k') # grafico de distribucion de cada producto
plt.ylabel('Unidades Totales', fontsize = 20) 
plt.xlabel('Productos', fontsize = 20) 
plt.title('Ventas de productos -Box', fontsize=20) # titulo
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python

sells_df = df.drop(['month_number', 'total_units', 'total_profit'], axis=1) # agrupacion de los datos
cols = sells_df.columns # columnas

mouse = sells_df.apply(lambda x: sum(x)) # ventas totales del año de cada producto



# grafica de torta
plt.pie(
    
        mouse, # datos
        labels =cols, # etiqueta de los datos
        autopct = '%1.1f%%' # porcentaje
    
       )

plt.title('Ventas totales del año pasado', fontsize=30) # titulo
plt.show()
```
